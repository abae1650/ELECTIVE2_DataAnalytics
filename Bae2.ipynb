```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

A.

```python
df = pd.read_csv("movie_metadata_cleaned.csv")
df.head(5)
```

```python
df.tail(5)
```

```python
df.count()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['gross'], df['director_facebook_likes'])
ax.set_xlabel('gross')
plt.title('Figure 1')
ax.set_ylabel('director_facebook_likes')
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['gross'], df['actor_1_facebook_likes'])
ax.set_xlabel('gross')
plt.title('Figure 2')
ax.set_ylabel('actor_1_facebook_likes')
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['gross'], df['num_critic_for_reviews'])
ax.set_xlabel('gross')
plt.title('Figure 3')
ax.set_ylabel('num_critic_for_reviews')
plt.grid()
plt.show()
```
